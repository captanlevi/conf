```python
from scapy.all import rdpcap, UDP
```

```python
def extract_first_8_bytes(pcap_file):

    def formatNumber(data):
        return "-".join(f"{byte:02X}" for byte in data)
    
    dct = dict()
    dtls_headers = dict()
    packets = rdpcap(pcap_file)  # Read pcapng file
    print(len(packets))
    for pkt in packets:
        if pkt.haslayer(UDP):  # Check if it's a UDP packet
            udp_payload = bytes(pkt[UDP].payload)  # Extract UDP payload
            
            if len(udp_payload) >= 13:  # Ensure payload is at least 8 bytes
                dtls_header = formatNumber(udp_payload[:13])
                dtls_headers[dtls_header] = 1 + dtls_headers.get(dtls_header, 0)

                first_8_bytes = formatNumber(udp_payload[13:13 + 8])  # Get first 8 bytes
                dct[first_8_bytes] = 1 + dct.get(first_8_bytes,0)
        else:
            print("non UDP packet found")
    
    return dct,dtls_headers
```

```python
count,dtls_headers = extract_first_8_bytes("zoomMeetCap.pcapng")
```
